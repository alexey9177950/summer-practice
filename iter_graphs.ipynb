```python
import matplotlib.pyplot as plt
import math
```

```python
file = open("data_iter_data.txt")
dim = list(map(int, file.readline().split()))
iter_n = list(map(int, file.readline().split()))

plt.figure(figsize = (10, 7))
plt.grid(True) # сетка
plt.xlabel("Количество итераций")
plt.ylabel("Погрешность (десятичный логарифм)")
plt.title("Метод простой итерации")
for n in dim:
    val = list(map(float, file.readline().split()))
    plt.plot(iter_n, list(map(lambda x : math.log10(x), val)), label = str(n))
plt.legend()
plt.show()

file = open("data_iter_2_data.txt")
dim = list(map(int, file.readline().split()))
iter_n = list(map(int, file.readline().split()))
plt.figure(figsize = (10, 7))
plt.grid(True) # сетка
plt.xlabel("Количество итераций")
plt.ylabel("Погрешность (десятичный логарифм)")
plt.title("Метод Зейделя")
for n in dim:
    val = list(map(float, file.readline().split()))
    plt.plot(iter_n, list(map(lambda x : math.log10(x), val)), label = str(n))
plt.legend()
plt.show()
```

```python
file = open("not_conv.txt")
iter_n = []
acc = []
for row in file:
    buf_1, buf_2 = row.split()
    iter_n.append(int(buf_1))
    acc.append(float(buf_2))

plt.figure(figsize = (10, 7))
plt.grid(True) # сетка
plt.plot(iter_n, acc)
plt.xlabel("Количество итераций")
plt.ylabel("Погрешность (десятичный логарифм)")
plt.title("Метод Зейделя")
plt.show()
```
