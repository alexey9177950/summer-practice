```python
import matplotlib.pyplot as plt
```

```python
dim = open("data/time_dim.txt").read().split()
dim = list(map(int, dim))
m_num = 4
res = []
labels = ["метод Гаусса", "выбор главного элемента", "выбор главного элемента 2", "метод вращений"]
for line in open("data/time.txt"):
    res.append(list(map(float, line.split())))
```

```python
plt.figure(figsize=(10, 6))
for i in range(m_num):
    plt.plot(dim, res[i], label = labels[i])
plt.xlabel("Размер матрицы") 
plt.ylabel("Время в секундах") 
plt.title('Время работы алгоритмов')
plt.grid(True) # сетка
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

for i in range(m_num):
    val = [i_y/ (i_x ** 3) for i_x, i_y in zip(dim, res[i])]
    plt.plot(dim, val, label = labels[i])
plt.xlabel("Размер матрицы") 
plt.title('Время работы алгоритмов (коэффициент перед $n^3$)')
plt.grid(True) # сетка
plt.legend(loc=2)
plt.show()
for i in range(m_num):
    print(res[i][-1] / dim[-1] ** 3)
```
